```python
def to_snake_case(text):
    text = str(text)
    text = text.lower
    text = text.replace(" ", "_")
    text = text.replace("/", "_")
    text = text.replace("-", "_")
    return text

df.columns = df.columns.apply(to_snake_case)
```
